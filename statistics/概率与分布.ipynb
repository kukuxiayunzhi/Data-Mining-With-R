## 概率分布

概率论与数理统计是研究随机现象统计规律性的一门学科. 对于一个具体的问题,通常归结为对一个随机变量或随机向量($X$)的取值及其取值的研究. 这就是随机变量的累积分布函概率的研究, 即对于事件$P(X\leq x)$的研究.这就是随机变量的累计分布函数($CDF$),记为$F(x)$.因此随机变量统计规律可以用累积分布函数来刻划. 
- 对于离散型随机变量(取值为有限或可列无限), 其统计规律通常转化为对分布律$f(x)=P(X=x)$的研究,它与分布函数的关系为$F(x)=$
$\sum^{}_{t \leq x} {P(X=t)}$;
- 而对于连续型随机变量(取值充满整个区间), 其统计规律通常转化为对概率密度函数$f(x)$的研究, 它与分布函数的关系为$F(x)=\int^{x}_{-\infty}{f(x)dx}$. 下面我们分离散与连续二种情况分别介绍它们的分布律或密度函数, 在此我们不加区分地使用$f(x)$.

### 离散分布的分布律

   #### 贝努利分布
   
- 意义: 一试验中有二个事件: 成功(记为1)与失败(记为0), 出现的概率
  是分别为$p$和$1-p$, 则一次试验(称为贝努里试验)成功的次数服从一
  个参数为p的贝努里分布.
- 分布律：
    $f(x | p)=p^{x}(1-p)^{1-x}, \quad x=0,1 \quad(0<p<1)$
- 数字特征：
    $\mathrm{E}(X)=p, \operatorname{Var}(X)=p(1-p)$
    
   #### 二项分布
   
- 意义：贝努里试验独立地重复n次, 则试验成功的次数服从一个参数
    为(n, p)的二项分布.
- 分布律:
    $f(x | n, p)=\left(\begin{array}{l}{n} \\ {p}\end{array}\right)   p^{x}(1-p)^{n-x}, \quad x=0,1, \cdots, n$
- 数字特征 :   $\mathrm{E}(X)=n p, \operatorname{Var}(X)=n p(1-p)$

   #### 多项分布
   
- 意义:一试验中有ｋ个事件$A_{i}, i=1,2, \cdots, k$，且$P\left(A_{i}\right)=p_{i}$$\left(0<p_{i}<1, \sum_{i=1}^{k} p_{i}=1\right)$.
  将次试验独立地重复ｎ次，则事件$A_{i}, i=1,2, \cdots, k$出现的次数服从一个参数为 $(n,p)$的多项分布，其中$\mathbf{P}=\left(p_{1}, p_{2}, \cdots, p_{k}\right)$.
- 分布律:
    $f\left(x_{1}, \cdots, x_{k} | n, \mathbf{p}\right)=\frac{n !}{x_{1} ! \cdots x_{k} !} p_{1}^{x_{1}} p_{2}^{x_{2}} \cdots p_{k}^{x_{k}}, \quad 0 \leqslant x_{i} \leqslant n, \sum_{i=1}^{k} x_{i}=n$
- 数字特征:  $\mathrm{E}\left(X_{i}\right)=n p, \operatorname{Var}\left(X_{i}\right)=n p(1-p), \operatorname{Cov}\left(X_{i}, X_{j}\right)=-n p_{i} p_{j}$

   #### 负二项分布
   
- 意义:贝努里试验独立地重复进行, 一直到出现k次成功时停止试验,
  则试验失败的次数服从一个参数为(k, p)的负二项分布.
- 分布律:$f(x | k, p)=\frac{\Gamma(k+x)}{\Gamma(k) \Gamma(x)} p^{k}(1-p)^{x}, \quad x=0,1, \cdots$
- 数字特征: $\mathrm{E}(X)=\frac{k(1-p)}{p}, \operatorname{Var}(X)=\frac{k(1-p)}{p^{2}}$

   #### 几何分布
   
- 意义:努里试验独立地重复进行, 一直到出现有成功出现时停止试
  验, 则试验失败的次数服从一个参数为p的几何分布.
- 分布律:$f(x | p)=p(1-p)^{x}, \quad x=0,1,2, \cdots$
- 数字特征:  $\mathrm{E}(X)=\frac{(1-p)}{p}, \operatorname{Var}(X)=\frac{(1-p)}{p^{2}}$

   #### 超几何分布
   
- 意义:从装有N个白球和M个黑球的罐子中不放回地取出k(<=M+N)个球, 则其中的白球数服从超几何分布
- 分布律:$f(x | N, M, k)=\frac{\left(\begin{array}{c}{N} \\ {x}\end{array}\right)\left(\begin{array}{c}{M} \\ {k-x}\end{array}\right)}{\left(\begin{array}{c}{N+M} \\ {k}\end{array}\right)}, \quad x=0,1,2, \cdots \min \{N, k\}$
- 数字特征: $\mathrm{E}(X)=\frac{(k N)}{N+M}, \operatorname{Var}(X)=\left(\frac{N+M-k}{N+M-1}\right) \frac{k N}{N+M}\left(1-\frac{N}{N+M}\right)$ 

   #### 泊松分布
   
- 意义:单位时间, 单位长度, 单位面积, 单位体积中发生某一事件的次数常可以用泊松(Poisson)分布来刻划, 例如某段高速公路上一年内的交通事故数和某办公室一天中收到的电话数可以认为近似服从泊松分布.
- 分布律:$f(x | \lambda)=\frac{\lambda^{x}}{x !} e^{-\lambda}, \quad x=1,2, \cdots$
- 数字特征: $\mathrm{E}(X)=\lambda, \operatorname{Var}(X)=\lambda$


### 几个例子

```r
#二项分布
par(mar = c(2.5,4,2,0.1),mfrow = c(2,3))
n <-c(8,8,8,5,10,30)
p <- c(0.1,0.3,0.5,0.2,0.2,0.2)
for(i in 1:length(p)){
    n1<-n[i]
    p1<-p[i]
    x<-0:n1
    px<-choose(n1,x)*p1^x*(1-p1)^(n1-x)
##绘制二项分布的概率分布
plot(x,
    type='n',
    xlim=c(0,n1),
    ylim=c(0,max(px)+0.025),
    ylab='P(x)',
    main=paste0('n=',n1,',p=',p1),
    cex.main=2)
rect(x-0.1,0,x+0.1,px,col = 'green')

##添加正态分布曲线，均值为np和方差为nqp
xv <- seq(-1,n1,length.out = 1000)
yv <- dnorm(xv,mean = n1*p1,sd=sqrt(n1*p1*(1-p1)))
lines(xv,yv,col = 'red',lwd = 1.5)
}
```

### 连续分布的密度函数
#### 贝塔分布
- 意义:在贝叶斯分析中, 贝塔分布常作为二项分布参数的共轭先验分
    布.
- 密度函数:$f(x | a, b)=\frac{1}{B(a, b)} x^{a-1}(1-x)^{b-1}, \quad 0<x<1 \quad(a, b>0)$
- 数字特征:$\mathrm{E}(X)=\frac{a}{a+b}, \operatorname{Var}(X)=\frac{a b}{(a+b)^{2}(a+b+1)}$
#### 均匀分布
- 意义:区间 $[a,b]$ 上随机投点对应的坐标服从 $[a,b]$ 上的均匀分布.
- 密度函数:$f(x | a, b)=\frac{1}{b-a}, \quad a \leqslant x \leqslant b$
- 数字特征:$\mathrm{E}(X)=\frac{a+b}{2}, \operatorname{Var}(X)=\frac{b^{2}-a^{2}}{12}$
#### 柯西分布
- 意义:柯西分布(又称为Lorentz分布)用于描述共振行为. 以一随机的
    角度投向X轴的水平距离服从柯西分布.
- 密度函数:$f(x | a, b)=\frac{1}{\pi b\left[1+\left(\frac{x-a}{b}\right)\right]}, \quad 0<x<1 \quad(a, b>0)$
- 数字特征:均值与方差不存在
#### 威布尔分布
- 意义:最为常用的寿命分布, 用来刻划滚珠轴承、电子元器件等产品
    的寿命.
- 密度函数:$f(x | a, b)=a b x^{b-1} e^{a x^{b}}, \quad x>0 \quad(a, b>0)$
- 数字特征:$\mathrm{E}(X)=\frac{\Gamma\left(1+\frac{1}{b}\right)}{a^{1 / b}}$$\operatorname{Var}(X)=\frac{\Gamma\left(1+\frac{2}{b}\right)}{a^{2 / b}}-\frac{\left\{\Gamma\left(1+\frac{1}{b}\right)\right\}^{2}}{a^{2 / b}}$
#### 指数分布
- 意义:泊 松 过 程 的 等 待 时 间 服 从 指 数 分 布. 形 状 参 数b=1的Weibull   分布为指数分布.
- 密度函数:$f(x | \lambda)=\lambda e^{-\lambda x}, \quad x>0 \quad(\lambda>0)$
- 数字特征:$\mathrm{E}(X)=\frac{1}{\lambda}, \operatorname{Var}(X)=\frac{1}{\lambda^{2}}$
#### 瑞利分布
- 意义:瑞利(Rayleigh)分布为Weibull分布的又一个特例: 它是参数
  为 $(1/(2b^2),2)$的Weibull分布.
- 密度函数:$f(x | b)=\frac{x}{b^{2}} \exp \left(-\frac{x^{2}}{2 b^{2}}\right)$
- 数字特征:$\mathrm{E}(X)=\sqrt{\frac{\pi}{2}} b, \operatorname{Var}(X)=\frac{4-\pi}{2} b^{2}$


#### 正态分布
- 意义:高斯分布是概率论与数理统计中最重要的一个分布. 中心极限
    定理表明, 一个变量如果是由大量微小的、独立的随机因素的叠加
    结果, 那么这个变量一定是正态变量. 因此许多随机变量可以用高
    斯分布表述或近似描述.
- 密度函数:$\begin{aligned} f(x | \mu, \sigma)=\frac{1}{\sqrt{2 \pi} \sigma} e^{-\frac{(x-\mu)^{2}}{2 \sigma^{2}}}, &-\infty<x<\infty \\(-\infty<\mu<\infty, \sigma>0) \end{aligned}$
- 数字特征:$\mathrm{E}(X)=\mu, \operatorname{Var}(X)=\sigma^{2}$
#### 对数正态分布
- 意义:$ln（ｘ）$服从参数为 $(μ,σ^2)$ 的正态分布, 则X服从参数为$(μ,σ^2)$的
  对数正态分布.
- 密度函数:$\begin{aligned} f(x | \mu, \sigma)=& \frac{1}{\sqrt{2 \pi} \sigma x} e^{-\frac{(\ln (x)-\mu)^{2}}{2 \sigma^{2}}}, \quad x>0 \\ &(-\infty<\mu<\infty, \sigma>0) \end{aligned}$
- 数字特征:$\mathrm{E}(X)=\exp \left\{\mu+\frac{1}{2} \sigma^{2}\right\}, \operatorname{Var}(X)=e^{\sigma^{2}}\left(e^{\sigma^{2}}-1\right) e^{2 \mu}$
#### 逆正态分布
- 意义:正态随机变量的倒数服从的分布
- 密度函数:$f(x | \mu, \lambda)=\sqrt{\frac{\lambda}{2 \pi x^{3}}} \exp -\frac{\lambda(x-\mu)}{2 \mu^{2} x} \quad(-\infty<\mu<\infty, \lambda>0)$
- 数字特征:$\mathrm{E}(X)=\mu, \operatorname{Var}(X)=\frac{\mu^{3}}{\lambda}$
#### 珈玛分布
- 意义:k个相互独立的参数为1/b的指数分布的和服从参数为$(k, b)$ 的
    珈玛分布
- 密度函数:$f(x | a, b)=\frac{1}{\Gamma(a) b^{a}} x^{a-1} e^{-x / b}$
- 数字特征:$\mathrm{E}(X)=a b, \operatorname{Var}(X)=a b^{2}$
#### 逆珈玛分布
- 意义:珈玛分布随机变量的倒数服从逆珈玛分布
- 密度函数:$f(x | a, b)=\frac{1}{\Gamma(a) b^{a}} x^{-(a+1)} e^{-1 /(b x)}, \quad x>0 \quad(a>0, b>0)$
- 数字特征:$\mathrm{E}(X)=\frac{1}{b(a-1)}(a>1), \operatorname{Var}(X)=\frac{1}{b^{2}(a-1)^{2}(a-2)}(a>2)$
#### 卡方分布
- 意义:n个独立正态随机变量的平方和服从自由度为n的卡方分布.
- 密度函数:$$f(x | n)=\frac{x^{n / 2-1} e^{-x / 2}}{2^{n / 2} \Gamma(n / 2)}, \quad x>0$$
- 数字特征:$\mathrm{E}(X)=n, \operatorname{Var}(X)=2 n \quad(n>2)$
#### 逆卡方分布
- 意义:卡方分布随机变量的倒数服从逆卡方分布

- 密度函数:
$$f(x | n)=\frac{x^{-(n / 2+1)} e^{-1 / 2 x}}{2^{n / 2} \Gamma(n / 2)}, \quad x>0$$

- 数字特征:$\mathrm{E}(X)=\frac{1}{n-2} \quad(n>2), \operatorname{Var}(X)=\frac{2}{(n-2)^{2}(n-4)} \quad(n>4)$
#### t分布
- 意义:随机变量X与Y独立, X服从标准正态分布, Y 服从自由度为n卡方分布, 则$T=\frac{X}{\sqrt{Y / n}}$服从自由度为n的t分布.

- 密度函数:$$
f(x | n)=\frac{\left(1+\frac{x^{2}}{n}\right)^{-(n+1) / 2}}{\sqrt{n B\left(\frac{1}{2}, \frac{n}{2}\right)}}
$$

- 数字特征:$\mathrm{E}(X)=0, \operatorname{Var}(X)=\frac{n}{n-2} \quad(n>2)$
#### F分布
- 意义:随机变量X与Y 独立, X服从自由度为n卡方分布, Y服从自由度为ｍ卡方分布，则$T=\frac{X / n}{Y / n}$服 从自由度为$（n, m）$的F 分布.
- 密度函数:$$f(x | n, m)=\frac{\left(\frac{n}{m}\right)^{n / 2} x^{n-2} / 2}{B\left(\frac{n}{2}, \frac{m}{2}\right)}\left(1+\frac{n}{m} x\right)^{-(n+m) / 2}$$
- 数字特征:$$\mathrm{E}(X)=\frac{m}{m-2} \quad(m>2), \operatorname{Var}(X)=\frac{2 m^{2}(n+m-2)}{n(m+2)} \quad(n>2)$$
#### logistic分布
- 意义:生 态 学 中 的 增 长 模 型 常 用logistic分 布 来 刻 划, 它 也 常 用
    于logistic回归中.
- 密度函数:$$
f(x | a, b)=\left[1+e^{-(x-a) / b}\right]^{-1}
$$
- 数字特征:$$
\mathrm{E}(X)=a, \operatorname{Var}(X)=\frac{\pi^{2}}{3} b^{2}
$$
#### Dirichlet分布
- 意义:在贝叶斯分析中可作为多项分布参数的共轭分布. Dirichlet分
    布的密度函数表示在已知k个竞争事件已经出现了$α_i-1$次条件下,
    它们出现的概率为$x_{i}, i=1,2, \cdots, k$的信念.
- 密度函数:$f\left(x_{1}, \cdots, x_{k} | \alpha\right)=\frac{1}{B(\alpha)} \prod_{i=1}^{k} x_{i}^{\alpha_{i}-1}, \quad x_{i}>0, \sum_{i=1}^{k} x_{i}=1 \quad\left(\alpha_{i}>0\right)$,其中$B(\alpha)=\frac{\prod_{i=1}^{k} \Gamma\left(\alpha_{i}\right)}{\Gamma\left(\sum_{i=1}^{k} \alpha_{i}\right)}$
- 数字特征:$\mathrm{E}(X)=\frac{\alpha_{i}}{\alpha_{0}}, \quad \operatorname{Var}(X)=\frac{\alpha_{i}\left(\alpha_{0}-\alpha_{i}\right)}{\alpha_{0}^{2}\left(\alpha_{0}+1\right)}$，
$\operatorname{Cov}\left(X_{i}, X_{j}\right)=-\frac{\alpha_{0} \alpha_{i}}{\alpha_{0}^{2}\left(\alpha_{0}+1\right)}$，其中$\alpha_{0}=\sum_{i=1}^{k} \alpha_{i}$
#### Pareto分布
- 意义:财富的分配的规则(称为Pareto规则)是大部分的财富(80%)被
    少数人(20%)的人拥有, 这可以较好地用Pareto分布来刻划.
- 密度函数:$f(x|a,b)=\frac{b}{a}\left(\frac{a}{x}\right)^{b+1}, \quad x>a \quad(b>0)$
- 数字特征:$\mathrm{E}(X)=\frac{a b}{b-1} \quad(b>1), \operatorname{Var}(X)=\frac{a^{2} b}{(b-1)^{2}(b-2)} \quad(b>2)$
#### 非中心分布
-  非中心的卡方分布:n个独立正态随机变量的平方和服从自由度为n，非中心参数为$\mu=\frac{\mu_{1}^{2}+\mu_{2}^{2}+\cdots, \mu_{n}^{2}}{\sigma^{2}}$的卡方分布.
-  非中心的ｔ分布:随机变量X与Y独立, X服从标准正态分布, Y 服从自由度为n卡方分布, 则$T=\frac{X}{\sqrt{Y / n}}$服从自由度为n、非中心参数为μ的t分布
-  非中心的Ｆ分布:随机变量X与Y 独立, X服从自由度为n卡方分布, Y服从自由度为ｍ卡方分布，则$T=\frac{X / n}{Y / n}$服 从自由度为$（n, m）$、非中心参数为μ的F 分布.

### 中心极限定理


- 正态分布在概率统计中起着至关重要的作用,其中的一个原因是当独立观察(试验)的样本容量n足够大时, 那么所观察的   随机变量近似服从正态分布
$$
\frac{\sum_{i=1}^{n} X_{i}-n \mu}{\sqrt{n} \sigma} \sim N(0,1) \quad(n \rightarrow \infty)
$$

```r
limite.central <- function(r=runif,distpar=c(0,1),m=.5,
                          s=1/sqrt(12),
                          n=c(1,3,10,30),N=1000){
    for(i in n){
        if(length(distpar)==2){
            x <- matrix(r(i*N,distpar[1],distpar[2]),nc=i)
        }
        else{
            x <- matrix(r(i*N,distpar),nc=i)
        }
        x <- (apply(x,1,sum) -i*m)/(sqrt(i)*s)
        hist(x,col='light blue',probability = T,main = paste("n=",i),
            ylim=c(0,max(.4,density(x)$y)))
        lines(density(x),col='red',lwd = 3)
        curve(dnorm(x),col='blue',lwd=3,lty=3,add = T)
        if(N>100){
            rug(sample(x,100))
        }
        else{
            rug(x)
        }
    }
}
```

```r
#二项分布
op <- par(mfrow = c(2,2))
limite.central(rbinom,distpar = c(10,0.1),m=1,s=0.9)
par(op)
```

```r
#泊松分布
op<-par(mfrow = c(2,2))
limite.central(rpois,distpar = 1,m = 1,s = 1,n = c(3,10,30,50))
par(op)
```
